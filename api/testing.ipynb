```python
def cleanPage(html_content: str) -> str:
    """
    Cleans the html content of the page
    """
    soup = BeautifulSoup(html_content, 'html.parser')
    for script in soup(["script", "style"]):
        script.extract()    # rip it out
    text = soup.get_text()
    return text

def cleanPage2(html_content: str) -> str:
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all the relevant elements (paragraphs and headings)
    allowed_tags = ['p', 'h1', 'h2', 'h3']
    filtered_elements = soup.find_all(allowed_tags)

    # Extract and concatenate the text within those elements
    cleaned_text = '\n'.join(element.get_text() for element in filtered_elements)
    return cleaned_text
```
